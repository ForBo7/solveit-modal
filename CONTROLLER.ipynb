- **双向同步**（notebook ↔ `.py` 模块）

---

**`01_modal` — Modal Sandbox 供应层**

这个模块负责在 Modal 上创建并配置一个带 SSH 访问的沙箱环境。流程很清晰：

1. `mk_app` 用 `App.lookup(create_if_missing=True)` 查找或创建一个**持久化的** Modal 应用（不用 `app.run()` 的 `with` 块，因为 exit 后隧道就断了）。
2. `mk_img` 从 Debian Slim 基础镜像出发，链式安装系统包（`openssh-server`）和 Python 包（`ipykernel`）。
3. `mk_sandbox` 在该镜像上启动一个 `sleep infinity` 的 Sandbox，暴露未加密的 22 端口给 SSH。
4. `get_tunnel` 从 Sandbox 拿到 22 端口对应的公网 host 和 port。
5. `get_pubkey` / `inject_pubkey` 把环境变量里的 SSH 公钥写进沙箱的 `authorized_keys`。
6. `start_ssh` 启动 SSH 服务并轮询确认运行。
7. `mk_ssh` 封装成一个 lambda，方便后面 `ssh('command')` 直接调用。
8. `verify_ssh` 连上去打印系统信息做验收。

**为什么这么设计？** 因为 Modal 的 `Sandbox` 是唯一能维持交互式长连接的产品形态（Function/Cls 不行），而用 unencrypted TCP 隧道直接把 22 端口捅出去，就不需要 bore.pub / ngrok 之类的外部代理——SSH 自带加密。

---

**`02_ipyfernel` — 远程 Kernel 连接层**

这个模块在已有 SSH 通道的沙箱里启动 IPython kernel，然后用 ipyfernel 把 solveit 接过去：

1. 先导入 `01_modal` 的所有函数，走完一遍完整的 sandbox 供应流程。
2. `start_remote_kernel` 通过 SSH 在远端执行 `python3 -m ipykernel_launcher -f <path>`，后台启动 kernel，然后轮询等 JSON 文件出现 + 确认日志里有 `To connect another client`。
3. `get_remote_python` 用 `which python3` 拿远端 Python 路径。
4. 最后调 `register_remote_kernel` 告诉 ipyfernel 远端 Python 在哪，再调 `connect_existing_kernel(f'{host}:{port}')` 把 solveit 的 kernel 切过去。

**为什么这么设计？** ipyfernel 原设计依赖 bore.pub 或 ngrok 做 TCP 代理，但这里的 Modal unencrypted tunnel 直接替代了那一步。`connect_existing_kernel(tunnel)` 接收的就是 `host:port`——正好对上了。
