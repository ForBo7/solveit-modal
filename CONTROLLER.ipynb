## 消息编辑工具速查

所有 `msg_*` 编辑工具都直接操作当前 dialog 的消息内容，返回 unified diff。

| 工具 | 用途 |
|---|---|
| `msg_replace_lines` | 替换连续行块（推荐首选） |
| `msg_str_replace` | 精准替换小段字符串 |
| `msg_strs_replace` | 一次性做多个小替换 |
| `msg_insert_line` | 在指定行后插入新内容 |
| `msg_del_lines` | 删除行范围 |

关键参数：
- `id` — 消息 ID（或逗号分隔的多个 ID，或 `'all'`）
- `start_line` / `end_line` — 1-based 行号
- `re_filter` — 像 vi 的 `g/pattern/` 一样只处理匹配的行
- `log_changed=True` — 插入一条变更记录

## 核心原则

1. **编辑前必须重新读取** — 历史对话中的截断结果不能依赖，每次编辑前先 `view_msg(id)` 拿到准确行号
2. **从后往前编辑** — 行号会因为前面的编辑而偏移，从消息末尾往开头改
3. **工具失败就停** — 不要猜参数重试，先搞清楚原因
4. **复杂编辑后验证** — 重新读取相关区域确认改动正确

## 典型工作流

```
view_msg(id) → 确定改动行 → msg_replace_lines(...) → 需要更多编辑？→ 重新读 → 重复
```

---

另外还有 `file_*` 系列工具（`file_str_replace`、`file_replace_lines` 等），操作的是普通文件而非 dialog 消息，用法类似。
